### Splitting The Data Into Train and Test

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

### Building A Model

### 1) By Using Support Vector Machine Algorithm

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.svm import SVC
```

```python
m_svm = SVC()
```

```python
m_svm.fit(X_train,y_train)
```

So Above We Fitted Support Vector Model On Training Dataset

```python
pred = m_svm.predict(X_test)
pred
```

```python
accuracy_score(y_test,pred)
```

### By Using Confusion Matrix

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test, pred))
```

```python
print("Classification Report")
print(classification_report(y_test, pred))
```

So Above We Saw The Classification Report

### 2) By Using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
from sklearn. metrics import accuracy_score, confusion_matrix
predictions= logmodel. predict(X_test)
```

```python
percentage= logmodel.score(X_test,y_test)
res=confusion_matrix(y_test,predictions)
print("confusion matrix")
print(res)
```

```python
print(logmodel.score(X_train,y_train))
```
